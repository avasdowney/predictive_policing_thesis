```python
import pandas as pd
import pygwalker as pyg
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

np.random.seed(1)
```

```python
df = pd.read_csv('../data/all_census_aif360_data.csv')
df.head()
```

```python
my_dict = {0:'Low', 1:'High'}
df = df.replace({'SSL SCORE':my_dict})
df
```

```python
ax = sns.histplot(x = df['Education_Rate'], hue=df['SSL SCORE'], multiple="dodge", stat='density', shrink=0.8, bins=[10, 20, 30, 40, 50, 60, 70, 80, 90], common_norm=False)
ax.spines["top"].set_visible(False), ax.spines["right"].set_visible(False)
for p in ax.patches:
    ax.annotate('{:.3f}'.format(p.get_height()), (p.get_x()+.05, p.get_height()+0.001), rotation = 75)
ax.set_xlabel('Education Rate')
ax.set_title('Density of SSL Score by Education Rate', y=1.10)
ax.xaxis.set_major_locator(mtick.MultipleLocator(10))
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
```
