```python
import pandas as pd
import numpy as np

np.random.seed(1)
```

```python
# read data
df = pd.read_csv('../data/Strategic_Subject_List_-_Historical.csv')
df = df[['SSL SCORE', 'PREDICTOR RAT AGE AT LATEST ARREST', 'PREDICTOR RAT VICTIM SHOOTING INCIDENTS', 'PREDICTOR RAT VICTIM BATTERY OR ASSAULT', 'PREDICTOR RAT ARRESTS VIOLENT OFFENSES', 'PREDICTOR RAT GANG AFFILIATION', 'PREDICTOR RAT NARCOTIC ARRESTS', 'PREDICTOR RAT TREND IN CRIMINAL ACTIVITY', 'PREDICTOR RAT UUW ARRESTS', 'CENSUS TRACT']]
df['SSL SCORE'] = np.where(df['SSL SCORE']<250 , 0, df['SSL SCORE']) # high risk = 1, low risk = 0
df['SSL SCORE'] = np.where(df['SSL SCORE']>=250 , 1, df['SSL SCORE'])
df = df.dropna()
df.describe()
```

```python
# One-hot encode the data using pandas get_dummies
df = pd.get_dummies(df)
```

```python
# with census tract
df.to_csv('../data/preprocessed_tract.csv', index=False)
df.head()
```

```python
# without census tract
df = df.drop(['CENSUS TRACT'], axis=1)
df.to_csv('../data/preprocessed_models_data.csv', index=False)
df.head()
```
